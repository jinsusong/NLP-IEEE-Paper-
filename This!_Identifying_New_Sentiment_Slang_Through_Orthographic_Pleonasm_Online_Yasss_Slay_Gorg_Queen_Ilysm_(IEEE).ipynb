예를 들어, 신조어를 식별하지 못하면 감정 사전에 너무 새로운 단어로 의견이 자주 표현되는 경우 어휘 감정 분석의 정확도가 떨어질 수 있습니다. 

이 기사는 소셜 웹에서 감정 관련 신조어를 식별하기 위해 맞춤법 pleonasm(용어법 : 불필요한 수식어를 붙이는 것)을 기반으로 하는 방법을 제안합니다. 

즉, 문자 반복 철자가 가장 다른 단어를 찾는 것입니다. 

이 방법은 대규모 소셜 웹 코퍼스에서 사전이 아닌 단어를 추출하고, 철자 표준화를 적용한 후 철자 변화 빈도가 낮은 순서로 단어를 순위 지정합니다. 

가장 많은 철자 변형이 있는 단어는 의미론적 컨텍스트에 대해 컨텍스트 내 키워드 분석됩니다. 

YouTube 인플루언서에 대한 댓글 모음에 적용한 이 방법은 전통적인 감정 단어, 느낌표 및 명사와 혼합된 slay 및 Early as a positive 용어와 같은 신조어를 찾습니다. 

orthographic(맞춤법) pleonasm은 원래 화자의 리듬과 음색을 표현하는 데 사용되었지만 발성하기 어려운 방식으로 이니셜에도 사용됩니다. 

따라서 이 방법은 일반적인 단어와 이니셜을 모두 포함하는 새로운 감성 속어를 식별하는 실용적인 방법입니다.

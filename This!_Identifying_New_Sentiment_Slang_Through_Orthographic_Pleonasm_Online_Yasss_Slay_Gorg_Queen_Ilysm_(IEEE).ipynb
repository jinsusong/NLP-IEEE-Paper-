#abstract

비공식 언어가 표준이고 청소년 속어가 일반적일 때 소셜 웹 텍스트의 자연어 처리에 신조어를 식별하는 것이 중요합니다. 

예를 들어, 신조어를 식별하지 못하면 감정 사전에 너무 새로운 단어로 의견이 자주 표현되는 경우 어휘 감정 분석의 정확도가 떨어질 수 있습니다. 

이 기사는 소셜 웹에서 감정 관련 신조어를 식별하기 위해 맞춤법 pleonasm(용어법 : 불필요한 수식어를 붙이는 것)을 기반으로 하는 방법을 제안합니다. 

즉, 문자 반복 철자가 가장 다른 단어를 찾는 것입니다. 

이 방법은 대규모 소셜 웹 코퍼스에서 사전이 아닌 단어를 추출하고, 철자 표준화를 적용한 후 철자 변화 빈도가 낮은 순서로 단어를 순위 지정합니다. 

가장 많은 철자 변형이 있는 단어는 의미론적 컨텍스트에 대해 컨텍스트 내 키워드 분석됩니다. 

YouTube 인플루언서에 대한 댓글 모음에 적용한 이 방법은 전통적인 감정 단어, 느낌표 및 명사와 혼합된 slay 및 Early as a positive 용어와 같은 신조어를 찾습니다. 

orthographic(맞춤법) pleonasm은 원래 화자의 리듬과 음색을 표현하는 데 사용되었지만 발성하기 어려운 방식으로 이니셜에도 사용됩니다. 

따라서 이 방법은 일반적인 단어와 이니셜을 모두 포함하는 새로운 감성 속어를 식별하는 실용적인 방법입니다.

이 기사는 맞춤법 pleonasm을 활용하여 소셜 웹 텍스트 세트에서 새로운 감정 속어를 식별하는 새로운 방법을 제안합니다. 

단어에 중복 문자가 추가되었습니다. 예를 들어, 누군가가 Facebook에서 wickiiiiiiid에 댓글을 달 수 있는데, 중복된 글자는 단어의 감정을 강조하는 기능을 합니다. 새로운 감정 관련 단어를 식별하기 위해 맞춤법 pleonasm을 사용하는 이면의 직관은 (a) 속어 철자가 속어 단어와 연관될 가능성이 높기 때문에 둘 다 비공식 언어 유형이므로 속어 단어는 광범위한 맞춤법 pleonasm으로 작성될 가능성이 있고 ( b) 감정 상태는 목소리 톤의 변화로 이어질 수 있으므로 감정 단어에 대한 음성 철자를 기대할 수 있습니다. leet(예: Lien et al. [26]), 맞춤법 pleonasm은 감정과 가장 자연스러운 연관성이 있는 것으로 보입니다.

1. 맞춤법 검사 없이 소셜 웹 텍스트 세트에서 모든 단어를 추출합니다.

# EXPERIMENT

영국 라이프스타일을 기반으로 하는 223명의 여성 인플루언서의 동영상에 대한 YouTube 댓글 세트에 철자법을 적용했습니다. 

이들은 무료 소프트웨어 Mozdeh를 통해 YouTube API로 2020년 11월–12월에 다운로드되었습니다. 

저자와 같은 국가의 세트는 속어 식별에 도움이 되며 라이프스타일 인플루언서 댓글은 특히 비공식적이고 감성이 풍부하므로 적절한 출처입니다. 

수집하기 쉽고 비교적 스팸이 없으며 날짜가 찍혀 있다는 추가적인 실용적인 이점이 있습니다. 

그러나 개별 YouTube 댓글 작성자는 때때로 다작일 수 있으며, 이로 인해 결과가 소수 사람의 작문 스타일로 왜곡될 수 있습니다. 

이를 방지하기 위해 각 댓글 작성자는 난수 생성기(Mozdeh에서)를 사용하여 한 달에 하나의 댓글로 제한되었습니다. 

이것은 많은 댓글 작성자에게 적당한 양의 추가 가중치를 부여하는 절충안입니다. 개인에 의한 지배를 더욱 방지하기 위해 각 단어를 가장 많이 사용하는 댓글 작성자는 후속 KWIC 프로세스(Mozdeh의 옵션) 동안 해당 용어의 단일 지배적 사용자인지 확인했습니다. 

이와 관련하여 우려할 만한 원인은 발견되지 않았습니다. 이 단계에서는 2007년 7월부터 2020년 12월까지 13,264,344개의 댓글이 생성되었습니다.

앞에서 설명한 대로 철자법이 적용되었으며 상위 100개 용어는 원본 YouTube 댓글과 함께 KWIC 접근 방식(Mozdeh의 일치자 사용)을 사용하여 속어 사용에 대해 작성자가 수동으로 확인했습니다. 

각 단어는 또한 실제로 철자 pleonasm에 대한 지원 컨텍스트 정보를 제공하기 위해 기본 기능에 대해 분류되었습니다.

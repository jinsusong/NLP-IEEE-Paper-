# A Sentiment Classification Method of Web Social Media Based on Multidimensional and Multilevel Modeling

### 다차원 및 다차원 모델링을 기반으로 한 웹 소셜 미디어의 감성 분류 방법

 소셜 미디어는 짧은 텍스트 길이, 비표준 언어, 그림과 텍스트의 특성을 가지고 있습니다. 이러한 특성으로 인해 웹 소셜 미디어의 감정 분류는 텍스트 컨텍스트 의미가 누락되는 문제에 직면하게 됩니다

1. 기존 연구는 주로 웹 소셜 미디어에서 언어 기호 정보를 마이닝하여 감정 분류를 구현합니다. 웹 소셜 미디어에서 이모티콘 기호와 구두점 기호에 포함된 감정 정보를 고려한 연구는 거의 없습니다. 

2. 기존 연구는 주로 문서 수준에서 감정 특징을 추출하고 텍스트 콘텐츠를 모델링하지만 단어, 문장, 문서와 같은 여러 수준에서 감정 특징을 추출하고 텍스트 콘텐츠를 모델링하는 것은 거의 고려하지 않습니다. 

MDMLSM(다차원 및 다단계 감성 모델링) 방법을 제안
1. 다차원 및 다단계 모델링을 기반으로 한 웹 소셜 미디어의 감성 분류 방법을 제안한다. Sina Weibo 및 Twitter 데이터 세트에 대한 실험은 우리의 방법이 감정 분류에서 경쟁 기준보다 5.86% 더 정확도를 향상시키는 것으로 나타났습니다.

언어 기호, 이모티콘 기호 및 구두점 기호의 3차원에서 다차원 감정 모델링(MDSM) 방법을 제안합니다. 다차원 감정 정보를 마이닝하여 웹 소셜 미디어에서 누락된 텍스트 컨텍스트 의미의 문제를 부분적으로 해결합니다.

우리는 단어, 문장, 문서의 세 가지 수준에서 MLSM(Multi-Level Sentiment Modeling) 방법을 제안합니다. 다단계 감정 정보를 마이닝하여 웹 소셜 미디어 텍스트에서 텍스트 컨텍스트 의미의 부족을 부분적으로 해결했습니다.

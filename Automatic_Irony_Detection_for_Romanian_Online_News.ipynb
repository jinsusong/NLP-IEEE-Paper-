1. 유니그램 빈도 행렬 계산
    : 유니그램 모델에서는 단어의 시퀀스를 고려한다기보다는 단어 셋(set)을 상정한다는 것이 더 정확한 표현입니다. 
    : 학습말뭉치에 등장한 각 단어 빈도를 세어서 전체 단어수로 나누어준 것입니다. 물론 확률의 총합은 1이 됩니다.

    1. 언어모델 
        - 정의 :   언어모델이란 단어 시퀀스에 대한 확률분포(probability distribution)를 가리킵니다. 언어모델은 m개 단어가 주어졌을 때 m개 단어 시퀀스가 나타날 확률, 즉 P(w1,w2,…,wm)을 할당(assign)합니다. 예컨대 다음과 같습니다
        - (1) P(Today is Wednesday)=0.001
        - (2) P(Today Wednesday is)=0.0000000001
        - 언어 모델 이점 : 자연언어에는 기본적으로 불확실성(uncertainties)이 존재합니다. 그런데 언어모델은 이러한 불확실성을 단어 시퀀스의 출현 확률로 정량화(quantify)할 수 있는 장점을 가집니다. 예컨대 언어모델의 힘을 빌리면 앞선 예제에서 (1)이 나타날 확률은 (2)보다 107배 크다고 ‘숫자’로 말할 수 있게 됩니다.
        
2. 용어 빈도 TF (Term Frequency)& 역 문서 빈도 IDF (Inverse Document Frequency) : 
    - 자연어 처리 기술에서 일반적으로 사용되는 두 가지 용어입니다. 단어 발생과 그 기여 또는 영향을 찾는 데 사용되거나 문서의 주어진 문장에서 중요성을 말할 수 있습니다. 이 기술은 감정 분류에 더 자주 사용됩니다
    - 용어 빈도는 1개 문서 안에서 특정 단어의 등장 빈도를 의미
    - IDF는 문서의 모든 주어진 문장에서 특정 단어의 발생 정도를 측정
    - TF-IDF는 모든 문서에 자주 등장하는 단어는 중요도가 낮다고 판단하며, 특정 문서에만 자주 등장하는 단어는 중요도가 높다고 판단합니다. TF-IDF값이 낮으면 중요도가 낮은 것이며, TF-IDF값이 크면 중요도가 큰 것

# METHODOLOGY

아이러니 감지 시스템은 텍스트에 아이러니 및 풍자와 관련된 다양한 유형의 감정이 포함되어 있는지 여부를 분류하기 위한 것입니다.

여기서 아이러니한 텍스트에 대한 수동 주석 프로세스는 힘들고 시간이 많이 소요되는 방법론이며 이 작업을 위해 더 작은 데이터 세트에서만 사용할 수 있습니다. .



기계 학습에서 일반적인 작업은 여러 데이터에서 학습하고 예측할 수 있는 알고리즘을 분석하고 구성하는 것입니다. 

따라서 이 접근 방식의 주요 구성 요소는 주석이 달린 루마니아어 뉴스 말뭉치( 표 I 참조)와 237,134단어를 포함하는 루마니아어 사전( 표 II 참조)입니다.), 

기계 학습 시스템에서 제거된 나머지 432,029개 단어가 제안된 방법에서 사용되었습니다. 

루마니아어에 대한 기계 학습 아이러니 탐지 접근 방식에서 사용하는 데이터 세트는 제안된 시스템의 성능에 직접적인 영향을 미치는 중요한 측면입니다. 

이 논문을 위해 2016년에서 2020년 사이에 발행된 25,841개의 아이러니하지 않은 뉴스 항목과 14,064개의 아이러니한 뉴스 항목을 포함하는 700개 이상의 루마니아 온라인 뉴스 게시자로부터 데이터를 수집했으며 루마니아의 한 공공 기관 직원이 수동으로 주석을 달았습니다.

이 논문에서 훈련 세트는 25,841개 항목에서 무작위로 추출된 10,000개의 비아이러니 및 아이러니 텍스트 중 75%를 포함하고 테스트 세트는 14,064개 항목에서 무작위로 추출한 10,000개의 아이러니 및 비아이러니 텍스트 중 25%를 포함합니다. 

제안된 시스템의 성능을 강조하고 검증하기 위해 이전에 언급한 분류기를 새로운 8,128개의 뉴스 항목 세트에 적용했습니다. 이는 검증 세트를 나타내는 4,064개의 아이러니한 항목과 4,064개의 비아이러니한 항목으로 훈련 과정에 포함되지 않습니다.

#CONCLUTION

자동 아이러니 감지 작업은 텍스트를 아이러니와 비아이러니로 분류하는 현상으로 간주되며 이전 설문 조사에서 설계한 주요 시스템은 지도 학습 시스템의 초안을 기반으로 합니다.

제안된 지도 접근 방식을 평가하기 위해 교육 데이터 세트를 시스템 성능의 주요 지표로 사용했으며 결과에 따라 가장 성능이 좋은 분류기는 Naive Bayes입니다.

또 다른 중요한 관찰은 Naive Bayes 분류기가 평가 데이터 세트에 대해 91% 이상의 좋은 결과를 얻었다는 것입니다. 

따라서 우리는 일반적으로 사용되는 단어를 기반으로 아이러니를 자동으로 감지하는 데 사용되는 특정 필터 방법을 추가하여 기존 알고리즘을 사용하여 아이러니 감지에 대한 접근 방식을 설명했습니다.

실제 응용 프로그램의 아이러니 감지 시스템은 사회적 관점에서 대통령 선거에 이르기까지 큰 영향을 미칠 수 있습니다. 

감정 분석의 준지도 학습 프레임워크 외에도 온라인에서 상당한 변화를 일으키거나 대기업이나 공공 기관의 중요한 결정에 영향을 미칠 수 있습니다.

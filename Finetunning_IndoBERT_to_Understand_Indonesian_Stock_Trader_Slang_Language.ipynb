B. 데이터 전처리
IndoBERT 모델 [21] 에 데이터를 공급하기 전에 모델 의 의도된 표준과 호환되도록 사전 처리되어야 합니다. 이 연구에서는 그림 4 와 같이 데이터 전처리의 4단계를 구현 했습니다.
